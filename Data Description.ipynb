Data Preprocessing

```python
df_data.dropna(subset=["SEVERITYCODE"], axis=0, inplace=True)
df_Analysis = df_data[['SEVERITYCODE', 'OBJECTID', 'WEATHER', 'ROADCOND', 'LIGHTCOND', 'SPEEDING', 'JUNCTIONTYPE', 'COLLISIONTYPE']]
```

```python
df_Analysis.head()
```

```python
# Replace n/a with unknown
df_Analysis['WEATHER'].fillna('unknown', inplace=True)
df_Analysis['ROADCOND'].fillna('unknown', inplace=True)
df_Analysis['LIGHTCOND'].fillna('unknown', inplace=True)
df_Analysis['SPEEDING'].fillna('unknown', inplace=True)
df_Analysis['JUNCTIONTYPE'].fillna('unknown', inplace=True)
df_Analysis['COLLISIONTYPE'].fillna('unknown', inplace=True)
```

```python
# Update categorical data
df_Analysis = pd.concat([
    df_Analysis, 
    pd.get_dummies(df_Analysis['WEATHER']), 
    pd.get_dummies(df_Analysis['ROADCOND']), 
    pd.get_dummies(df_Analysis['LIGHTCOND']), 
    pd.get_dummies(df_Analysis['SPEEDING']),
    pd.get_dummies(df_Analysis['JUNCTIONTYPE']),
    pd.get_dummies(df_Analysis['COLLISIONTYPE'])
], axis=1)
```
